```python


# Load the image from the URL using skimage's imread
image_url = "https://github.com/pythonai170624/04.11.2024/blob/main/jinny.jpg?raw=true"
image = io.imread(image_url)

# 2. Display the original image
plt.imshow(image)
plt.axis('off')  # Turn off axis labels
plt.show()

# 3a. Print the shape (size) of the image
print("Image size:", image.shape)

# 3b. Rotate the image 90 degrees using numpy's transpose (switch width and height)
rotated_image = np.transpose(image, (1, 0, 2))  # Transpose height, width
plt.imshow(rotated_image)
plt.axis('off')
plt.show()

# 3c. Rotate the image 120 degrees using PIL
# Convert the numpy array to a PIL Image using Image.fromarray
pil_image = Image.fromarray(image)

# Rotate the image 120 degrees
rotated_pil_image = pil_image.rotate(120)

# Display the rotated image
rotated_pil_image.show()

# 3d. Print the first 50 columns of the image (and all rows)
first_50_columns = image[:, :50, :]
print("First 50 columns of the image:")
plt.imshow(first_50_columns)
plt.axis('off')
plt.show()
```

```python
# Set the random seed for reproducibility
np.random.seed(80)
```

```python
# Generate a 4x4 array of random integers between 1 and 100
array = np.random.randint(1, 101, size=(4, 4))
```

```python
# a. Print the sum of each column and each row
print("Sum of each column:")
print(np.sum(array, axis=0))  # Sum along columns (axis 0)
print("\nSum of each row:")
print(np.sum(array, axis=1))  # Sum along rows (axis 1
```

```python

# b. Print the maximum of each column and each row
print("\nMaximum of each column:")
print(np.max(array, axis=0))  # Max along columns (axis 0)
print("\nMaximum of each row:")
print(np.max(array, axis=1))  # Max along rows (axis 1)
```

```python
# c. Print the minimum of each column and each row
print("\nMinimum of each column:")
print(np.min(array, axis=0))  # Min along columns (axis 0)
print("\nMinimum of each row:")
print(np.min(array, axis=1))  # Min along rows (axis 1)
```

```python
# d. Print the average of each column and each row
print("\nAverage of each column:")
print(np.mean(array, axis=0))  # Mean along columns (axis 0)
print("\nAverage of each row:")
print(np.mean(array, axis=1))  # Mean along rows (axis 1)
```

```python
# e. Print all numbers greater than 50 (use a condition inside square brackets)
print("\nNumbers greater than 50:")
print(array[array > 50])  # Select all numbers greater than 50
```

**1. Median (חציון):**<br> 
The median is the middle value in a set of numbers when they are arranged in order (either ascending or descending). If there is an even number of data points, the median is the average of the two middle values.<br>
Example: Consider the dataset: 3, 5, 1, 8, 7

Step 1: Arrange the numbers in ascending order: 1, 3, 5, 7, 8
Step 2: The median is the middle value: 5.
If the dataset had an even number of values (e.g., 1, 3, 5, 7), the median would be the average of the two middle values:

(3 + 5) / 2 = 4.<br>
